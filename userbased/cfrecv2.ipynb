```python
import pandas as pd
movies = pd.read_csv("movies.csv",encoding="Latin1")
Ratings = pd.read_csv("ratings.csv")
# Tags = pd.read_csv("tags.csv",encoding="Latin1")
movies.head()
```

```python
ratings = pd.merge(movies,Ratings)
print(ratings.shape)
ratings.head()
UserRatings = ratings.pivot_table(index=['userId'],columns=['title'],values='rating').fillna(2.5, axis=1)
UserRatings.head()
print("Before: ",UserRatings.shape)
# UserRatings = UserRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)
#userRatings.fillna(0, inplace=True)
# print("After: ",UserRatings.shape)
```

```python
UserRatings
```

```python
corrMatrix = UserRatings.corr(method='pearson')
# print(corrMatrix.head(10))
import pickle
# pickle.dump(corrMatrix, open('corrMatrix.pkl', 'wb'))
def get_similar(movie_name,rating):
    similar_ratings = corrMatrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
input_movies= {"Interstellar (2014)":5, "Sherlock Holmes: A Game of Shadows (2011)":5, "Inception (2010)": 4}
similar_movies = pd.DataFrame()
for movie,rating in input_movies.items():
    similar_movies = similar_movies.append(get_similar(movie,rating),ignore_index = True)
similar_movies.head(10)
x= []
xz = similar_movies.sum()
for y in similar_movies.sum().sort_values(ascending=False).head(20).index:
    if y not in input_movies:
        x.append(y)
# print(type(xz))
x
```
